#**    Calculated Fields, Parameters & LOD Expressions**

Question 1 : What is a Calculated Field and why is it important in business dashboards?

Ans: A Calculated Field in Tableau is a custom field created using formulas to derive new data from existing data.

Importance:

Helps create meaningful metrics (e.g., Profit Ratio, Delivery Status)
Enables deeper analysis without changing source data
Supports business decision-making with custom KPIs
Automates calculations inside dashboards.

Question 2 : Why can’t parameters change visuals directly in Tableau?

Ans: Parameters do not directly control visuals because:
They are independent variables, not linked to data automatically
They only store values (like numbers, text)
To affect visuals, parameters must be used with:
Calculated fields
Filters
Reference lines

Question 3 : When should EXCLUDE LOD be used?

Ans: Use EXCLUDE LOD when:
You want to remove a dimension from the view calculation
You need aggregated results at a higher level than what's shown
 Example:
View shows Category + Product
You want calculation at Category level only
 → Use EXCLUDE Product.

Task 3: Top N Products Analysis (Parameter) Business Need.
Ans: Step 1: Create Parameter
Name: Top N
Data Type: Integer
Values: 5, 10, 15 (or any).

Step 2: Create Rank Calculation
RANK(SUM([Sales])).

Step 3: Create Filter Field
RANK(SUM([Sales])) <= [Top N].

Task 4: City-Level Benchmarking (FIXED LOD)
Calculated Field:
{ FIXED [City] : AVG([Sales]) }
This calculates average sales per city, independent of filters.

Task 5: Product-Level Impact (INCLUDE LOD)
Calculated Field:
{ INCLUDE [Product_Name] : AVG([Sales]) }
 This:
Calculates average sales at product level
Still works when viewing Category or Region level
